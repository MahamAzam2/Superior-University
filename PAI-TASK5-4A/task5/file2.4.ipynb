```python
#1- draw a line 

import cv2 
  
path = r'C:/Users/LAPTOPMART/Desktop/venv/darw.png'
 
image = cv2.imread(path)

window_name = 'Image'

start_point = (0, 0)

end_point = (250, 250)

color = (0, 255, 0)
thickness = 9

image = cv2.line(image, start_point, end_point, color, thickness)
cv2.imshow(window_name, image) 
```

```python
# 2- draw arrow line

import cv2 
path = r'C:/Users/LAPTOPMART/Desktop/venv/arrow.png'

image = cv2.imread(path) 

window_name = 'Image'

start_point = (0, 0) 
end_point = (200, 200) 
color = (0, 255, 0) 
thickness = 9

image = cv2.arrowedLine(image, start_point, end_point, 
									color, thickness) 
cv2.imshow(window_name, image) 
```

```python
#3- draw ellipse
import cv2  
    

path = r'C:/Users/LAPTOPMART/Desktop/venv/ellipsecircle.png'
    
image = cv2.imread(path) 

window_name = 'image'
   
center_coordinates = (120, 100) 
  
axesLength = (100, 50) 
  
angle = 0
  
startAngle = 0
  
endAngle = 360
color = (0, 0, 255) 
   
thickness = 5
image = cv2.ellipse(image, center_coordinates, axesLength, 
           angle, startAngle, endAngle, color, thickness) 
   
# Displaying the image  
cv2.imshow(window_name, image)  
```

```python
#4 draw a circle
# Python program to explain cv2.circle() method  
    
# importing cv2  
import cv2  
    
# path  
path = r'C:/Users/LAPTOPMART/Desktop/venv/Crcle.png'
    

image = cv2.imread(path) 

window_name = 'Image'
   
center_coordinates = (120, 50) 
radius = 20
color = (255, 0, 0) 
   
thickness = 2

image = cv2.circle(image, center_coordinates, radius, color, thickness) 
   
# Displaying the image  
cv2.imshow(window_name, image) 
```
