# Data Pre-Processing

```python
df.info()
```

```python
def fillNaObject(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ["diabetes","serum_creatinine","smoking","DEATH_EVENT"]
fillNaObject(columns)
```

```python
def fillNaInt(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])
columns =["age","anaemia","creatinine_phosphokinase","ejection_fraction"]
fillNaInt(columns)
```

```python
def fillNaFloat(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())
columns = ["high_blood_pressure","platelets","serum_sodium","sex","smoking","time"]
fillNaFloat(columns)
```

```python
df.info()
```

```python
def convertFloattoInt(cols):
    for i in cols:
        df[i] = df[i].astype('int64')

columns = ["high_blood_pressure","platelets","serum_sodium","sex","smoking","time","age","anaemia","creatinine_phosphokinase","ejection_fraction"]
convertFloattoInt(columns)
```

```python
df.info()
```

```python
df.drop('age', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
df_encoded = df.copy()
```

```python
def convertObjtoInt(cols):
    for i in cols:
        datalabelEncoder = LabelEncoder()
        df_encoded[i] = datalabelEncoder.fit_transform(df[i])

columns = ['anaemia','sex']
convertObjtoInt(columns)
```

```python
df_encoded.info()
```

```python
df_encoded.to_csv(r'encoded-data.csv', index = False, header = True)
```

# Train test- split

```python
trainData, testData = train_test_split(df_encoded, test_size=0.2, shuffle=False)
```

```python
print(trainData.shape, testData.shape)
```

```python
trainData.iloc[:,1:]
```

```python
# train_x = df_encoded.drop('anaemia')
train_x = trainData.iloc[:,1:]
train_x.head(2)
```

```python
train_y = trainData.iloc[:, 0]
train_y.head(2)
```

```python
test_x = testData.iloc[:,1:]
test_x.head(2)
```

```python
test_y = testData.iloc[:, 0]
test_y.head(2)
```

# Model Training using SVC

```python
print("-- Training using SVC on Training Data --")
print("-- Parameters & Values: ", end='')

# model_svc = SVC(gamma='auto', random_state=0)
# model_svc.fit(train_x, np.ravel(train_y))

model_svc = SVC()
model_svc.fit(train_x, train_y)

print(model_svc)
```

```python
# Saving Trained Model
pickle.dump(model_svc, open('model_svc.pkl', 'wb'))
```

```python
# Load saved Model
model = pickle.load(open('model_svc.pkl', 'rb'))
```

```python
model_predictions = model.predict(test_x)
```

```python
testdata_predict = testData.copy(deep=True)
pd.options.mode.chained_assignment = None

testdata_predict['Prediction'] = model_predictions
```

```python
model_accuracy_score = accuracy_score(testdata_predict['DEATH_EVENT'], testdata_predict['Prediction'])

print("-- Model Accuracy Score: ", end='')
print(round(model_accuracy_score,3))
```
